```python
Ans: Escape characters are the characters we want insert in a code for our reference or any other characters 
     which is of no literal meaning, they are usually started with backslash (\).
    
    Example-1: "The King said "WE WON""
        If we execute example-1, we will get error. On the other hand refer example-2
    
    Example-2: "The King said \"WE WON\""
        This will result in output: 'The King said "WE WON"'       
```

```python
6. What are the values of the given expressions?
'Hello, world!'[1]
'Hello, world!'[0:5]
'Hello, world!'[:5]
'Hello, world!'[3:]
```

```python
Ans: 
'Hello, world!'[1]    = 'e' 
'Hello, world!'[0:5]  = 'Hello'
'Hello, world!'[:5]   = 'Hello'
'Hello, world!'[3:]   = 'lo world!'
```

```python
8. What are the values of the following expressions?
'Remember, remember, the fifth of July.'.split()
'-'.join('There can only one.'.split())
```

```python
Ans: 
'Remember, remember, the fifth of July.'.split()  = ['Remember,', 'remember,', 'the', 'fifth', 'of', 'July.']
'-'.join('There can only one.'.split())           = 'There-can-only-one.'
```
