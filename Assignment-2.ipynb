```python
8. Write code that prints Hello if 1 is stored in spam, prints Howdy if 2 is stored in spam, and prints Greetings! if anything else is stored in spam.
```

```python
Ans:
def hello(spam):
    """prints Hello if 1 is stored in spam, prints Howdy if 2 is stored in spam, and prints Greetings! if anything else is stored in spam"""
    if spam==1:
        return ("Hello")
    elif spam==2:
        return ("Howdy")
    else:
        return ("Greetings!")
hello(5)
```

```python
12. Write a short program that prints the numbers 1 to 10 using a for loop. Then write an equivalent program that prints the numbers 1 to 10 using a while loop.
```

```python
Ans:
a, using "for"
def tennumbersfor():
    """This function prints numbers 1 to 10 using for loop"""
    list=[]
    for i in range(1,11):
        list.append(i)
    return list
tennumbersfor()

b,using "while"
def tennumberswhie():
    """This function prints numbers 1 to 10 using while loop"""
    list=[]
    i=1
    while(i<=10):
        list.append(i)
        i=i+1
    return list
tennumberswhie()
```
